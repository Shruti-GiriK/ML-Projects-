```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['area_type']=encoder.fit_transform(df['area_type'])
df['availability']=encoder.fit_transform(df['availability'])
df['location']=encoder.fit_transform(df['location'])
df['society']=encoder.fit_transform(df['society'])
df['size']=encoder.fit_transform(df['size'])
```

```python
df['total_sqft'].unique()
```

```python
import pandas as pd

def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    if tokens[0].isnumeric():  # Check if the token is numeric
        return float(tokens[0])
    else:
        return None  # Handle non-numeric cases

df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)
# Replace NaN values with 0 before converting to integers
df['total_sqft'] = df['total_sqft'].fillna(0).astype(int)
```

# Removing the outliers

```python
def remove_outliers_iqr(df,exclude_columns):
    for col in df.columns:
        if col== exclude_columns:
            continue
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        upper_bound = Q3 + 1.5 * IQR
        lower_bound = Q1 - 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
df=remove_outliers_iqr(df,exclude_columns='price')
```

# Prediction

```python
x=df.drop(['price'],axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
y_pred=model.predict(xtest)
y_pred
```

```python
model.score(xtest,ytest)
```
