Label Encoding into numerical values

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['person_gender']=encoder.fit_transform(data['person_gender'])

data['person_home_ownership']=encoder.fit_transform(data['person_home_ownership'])
data['loan_intent']=encoder.fit_transform(data['loan_intent'])
data['previous_loan_defaults_on_file']=encoder.fit_transform(data['previous_loan_defaults_on_file'])
```

Rerieving unique values

```python
data['person_education'].unique()
```

replace because its ordinal data

```python

data['person_education'].replace(['High School','Associate','Bachelor','Master','Doctorate'],[1,2,3,4,5],inplace=True)
```

## Prediction

```python
price_corr=data.corr()['loan_status']
price_corr=price_corr.drop('loan_status') #removes the correlation of b/n loan status
sns.heatmap(price_corr.to_frame(),annot=True,fmt='.2f',cmap='coolwarm',
            cbar=True,annot_kws={'size':10})
plt.title('Correlation between loan_status and other feature')
plt.show()
```

Setting x and y variables

```python
x=data.drop(['person_gender','person_age','person_education','person_income',
             'person_emp_exp','loan_amnt','loan_intent','cb_person_cred_hist_length','credit_score'],axis=1)
#x=data.drop(['person_gender','person_age','person_education','loan_amnt',
             #'person_emp_exp','loan_intent','cb_person_cred_hist_length','credit_score'],axis=1)
#x=data.drop('loan_status',axis=1)
y=data['loan_status']
```

train_test_split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

Training and finding the score of the model

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
y_pred=model.predict(xtest)
y_pred
```
